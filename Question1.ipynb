```python
#Question 1:

import numpy as np
import pandas as pd
from IPython.display import Markdown,display
customers=pd.read_csv('customers.csv')
items=pd.read_csv('Items.csv')
orders=pd.read_csv('orders.csv')
order_item=pd.read_csv('order_item.csv')
#copy all
customers_df=customers.copy()
items_df=items.copy()
orders_df=orders.copy()
order_item_df=order_item.copy()
```

```python
def printmd(message,data):
    text="**<u>"+message+"</u>**"
    display(Markdown(text),data)
```

```python
# research: Print for each table (in one operation per table) the column names, types, and number of null values (and more)
def df_information(df,df_name):
    info_df=pd.DataFrame({
        'Columns':df.columns,
        'df_types':df.dtypes.values,
        'na_counts':df.isnull().sum().values
    })
    printmd(f"information on data frame name '{df_name}':",info_df)
df_information(customers_df,"customers")
df_information(items_df,"items")
df_information(orders_df,"orders")
df_information(order_item_df,"order_item")
```

```python
# Print for each table (in one operation per table) the column names, minimum values, maximum values, etc.
def df_information_stat(df,df_name):
    printmd(f"Columns on data frame name '{df_name}':",df.columns)
    printmd(f"stat on data frame name '{df_name}':",df.select_dtypes(include='number').fillna(0).describe(include='all'))
df_information_stat(customers_df,"customers")
df_information_stat(items_df,"items")
df_information_stat(orders_df,"orders")
df_information_stat(order_item_df,"order_item")
```
